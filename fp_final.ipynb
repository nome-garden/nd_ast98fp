```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
```

# Importing DEBCat data from Vizier catalog

```python
from astroquery.vizier import Vizier
Vizier.ROW_LIMIT = -1
catalogs = Vizier.get_catalogs(["DEBCat"])

#https://astroquery.readthedocs.io/en/latest/vizier/vizier.html
```

```python
debcat = catalogs[0].to_pandas()
```

# Accessing stellar mass information from DEBCat
And combining into one dataframe for convenience

```python
filtered=debcat[['Name','logM1', 'logM2']] #the only relevant data from this dataset are the masses
mass1=debcat['logM1']
```

```python
A = list(debcat['logM1'])
B= list(debcat['logM2'])
C=A+B

protoallmass = pd.DataFrame(C,columns=['M'])
allmass=protoallmass['M']
```

# Creating the scatter plots

```python
#scatterplot with random-generated test data

import random
rand_list=[]
n=390
for i in range(n):
    rand_list.append(random.randint(1,25)) #creates a list of random solar masses from 1-25

protorandomdata = pd.DataFrame(rand_list,columns=['test solar mass'])
randomgendata = protorandomdata['test solar mass']
print(randomgendata)

#https://www.geeksforgeeks.org/generating-random-number-list-in-python/

testplot=sns.scatterplot(data=protorandomdata,x=randomgendata,y=(10/randomgendata**2.5))
testplot.set(xlabel="solar masses (M)",ylabel="lifespan in billions of years (T)",title="stellar mass-lifespan relationship (test)")
```

```python
#test scatter plot with only M1 values

M1T_plot=sns.scatterplot(data=filtered,x=10**mass1,y=(10/(10**mass1)**2.5))
plt.yscale("log")
M1T_plot.set(xlabel="solar masses",ylabel="lifespan in billions of years",title="stellar lifespan-mass relationship, M1 only")
```

```python
#scatterplot with M1 and M2 values

MT_plot=sns.scatterplot(data=protoallmass,x=10**allmass,y=(10/(10**allmass)**2.5))
plt.yscale("log")
MT_plot.set(xlabel="solar masses (M)",ylabel="lifespan in billions of years (T)",title="stellar mass-lifespan relationship")
```

# Best Fit Curve

```python
from sklearn.linear_model import LinearRegression

x_log = np.array((np.log(10**allmass)))
y_log = np.array((np.log((10/(10**allmass)**2.5))))  # This should now work without resulting in NaN values

plt.subplot(1, 2, 2)
plt.scatter(x_log, y_log, color='red')
plt.title('Transformed Linearized Dataset')
plt.xlabel('log(solar masses)')
plt.ylabel('log(lifespan in years)')

#linear regression
model = LinearRegression()
model.fit(x_log.reshape(-1, 1), y_log)  # Reshape x_log for sklearn
slope = model.coef_[0]
intercept = model.intercept_
print(f"Linear regression on transformed data: y = {slope:.2f}x + {intercept:.2f}")

# Plotting the regression line
y_pred = model.predict(x_log.reshape(-1, 1))
plt.plot(x_log, y_pred, color='green', label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.legend()

plt.show()

#https://dagster.io/glossary/data-linearizing
```
